###4 Make a pie chart for all car makers

```python
import plotly.express as px

Car_maker = ['audi','alfa-romero']
val_counts = [4,7]
#pie figure draw
fig = px.pie(values=val_counts,names=Car_maker, #assigning value= count value as we wook [508,409] & #names=will be returned by the Hdislabel function
             color=Car_maker, #labeled colour to pie chart
             color_discrete_map={'Audi':'#FF4500',  #adding colour to pie chart
                                 'Alfa-Romero':'#FFFF00'},
             title='Car_Maker Count') #top title to pie figure

fig.show()
```
